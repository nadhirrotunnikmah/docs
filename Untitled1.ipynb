```python
import numpy as np

# Fungsi untuk menghitung jarak antara titik
def hitung_jarak(p1, p2):
    return np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

# Titik-titik kota
titik_kota = {
    'A': (0, 0),
    'A1': (1, 2),
    'A2': (3, 5),
    'A3': (4, 7),
    'A4': (2, 4),
    'A5': (5, 1),
    'A6': (6, 3),
    'A7': (8, 6),
    'A8': (9, 8),
    'A9': (6, 8)
}

# Fungsi untuk menghitung fitness (biaya) dari rute
def hitung_fitness(rute):
    fitness = 0
    for i in range(len(rute) - 1):
        fitness += hitung_jarak(titik_kota[rute[i]], titik_kota[rute[i+1]])
    return fitness

# Pembentukan populasi awal untuk 3 rute yang telah ditentukan
def populasi_awal(jumlah_populasi, titik_kota):
    populasi = []
    titik = list(titik_kota.keys())
    for _ in range(jumlah_populasi):
        rute1 = ['A1', 'A2', 'A3', 'A4', 'A']
        rute2 = ['A1', 'A5', 'A6', 'A4', 'A']
        rute3 = ['A1', 'A7', 'A8', 'A9', 'A']
        np.random.shuffle(rute1)
        np.random.shuffle(rute2)
        np.random.shuffle(rute3)
        populasi.append([rute1.copy(), rute2.copy(), rute3.copy()])
    return populasi

# Perhitungan nilai fitness untuk setiap rute pada populasi
def nilai_fitness_populasi(populasi):
    fitness_pop = []
    for rutes in populasi:
        fitness_rutes = []
        for rute in rutes:
            fitness = hitung_fitness(rute)
            fitness_rutes.append(1 / fitness)  # Menggunakan invers dari fitness sebagai nilai fitness
        fitness_pop.append(fitness_rutes)
    return fitness_pop

# Mencari nilai probabilitas komulatif untuk seleksi
def probabilitas_komulatif(fitness_pop):
    total_fitness = np.sum(fitness_pop)
    prob_komulatif = [fitness / total_fitness for fitness in fitness_pop]
    prob_komulatif = np.cumsum(prob_komulatif)
    return prob_komulatif

# Algoritma genetika
def algoritma_genetika(jumlah_populasi, generasi):
    populasi = populasi_awal(jumlah_populasi, titik_kota)
    for i in range(generasi + 1):  # Memperbarui perulangan hingga generasi + 1
        fitness_pop = nilai_fitness_populasi(populasi)
        prob_komulatif = probabilitas_komulatif(fitness_pop)

        # Seleksi
        # Lakukan proses seleksi untuk generasi berikutnya

        # Crossover
        # Lakukan proses crossover untuk mendapatkan individu baru

        # Mutasi
        # Lakukan proses mutasi pada individu hasil crossover

        # Update populasi
        # Gabungkan individu lama dan baru untuk membentuk populasi berikutnya

    # Mengambil rute terbaik (dengan nilai probabilitas terkecil)
    fitness_pop = nilai_fitness_populasi(populasi)
    best_individual_index = np.unravel_index(np.argmax(fitness_pop), np.array(fitness_pop).shape)
    best_route = populasi[best_individual_index[0]][best_individual_index[1]]
    return best_route

# Contoh penggunaan algoritma genetika
jumlah_populasi = 10
generasi = 0  # Nilai generasi dimulai dari 0
solusi_akhir = algoritma_genetika(jumlah_populasi, generasi)
print("Solusi akhir:", solusi_akhir)
```
